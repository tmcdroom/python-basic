#### スライス処理　　:  （セミコロンで範囲指定ができる）

#### 便利な小技　飛び飛びで値ほしい　　[start:stop:step]

```python
print(e[1:9:3]) # index 1から9まで　2つ飛ばしで拾う step を省略すると　step = 1 扱いになる
print(e[1:9:1])
print(e[1:9:])
print(e[1:9])
```

```python
print(e[0:10:2])  # 始点終点は省略できるので 以下は全部等価です
print(e[0::2])
print(e[:10:2])
print(e[::2])
```

#### 便利な小技の応用　配列を逆順に抜き出したい

```python
print(e[::-1])   #始点と終点を省略している　＆　step = -1 にしている　⇒ stepが負の場合は　終点 : 始点 : ステップになるので注意
```

#### 参考    始点終点を省略しない　＆　step が負の場合の時だけ　以下の条件になるらしい　覚えなくていいです 
#### [ i : j : k ] としたとき
#### k > 0 のとき「 i 番目から j - 1 番目の要素を k 個ごとに取り出す（i < j）」
#### k < 0 のとき「 i 番目から j + 1 番目の要素を k 個ごとに取り出す（i > j）」

```python
print(e[10:0:-1]) # 上記ルールにより　これはダメ
print(e[-1:-11:-1]) # これならいいけど　おとなしく　::-1 を使いましょう
```
